## Import Necessary Libraries

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Image

```python
image=cv2.imread('pk.jpeg')
#let us see the image
plt.imshow(image)
```

As OpenCV loads an image in BGR format we need to convert our RBG format image to display it in true colours.

```python
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#Now let's see the image
plt.imshow(image)
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
#let us see the image again
plt.imshow(gray_image, cmap='gray')
```

## Haar cascade files

OpenCV comes with a bunch of pre-trained classsifiers.There are classifiers for smile, eyes,face etc.These were present in the form of xml files and are located in opencv/data/haarcascades/ folder. Download the xml files and place them in data folder in the same working directory as the jupyter notebook.

### Loading Frontal Face Classifier

```python
face_haar_cascade = cv2.CascadeClassifier('data/haarcascades/haarcascade_frontalface_alt2.xml')
```

## Face detection

We are using detectMultiscale module of the classifier.This function will return the co-ordinates of x and y position plus height and width of the detected faces as [x,y,w,h].

```python
total_faces = face_haar_cascade.detectMultiScale(gray_image, scaleFactor = 1.2, minNeighbors = 5)#scaleFactor for compensating images and minNeighbours=5 for Front Face.
# Let us print the count of faces 
print('Faces found: ', len(total_faces))
```

After getting all face coordinates, we have to bound the faces with a rectangle by looping over them.
we will be drawing a <span style="color:green"> Green </span></center> rectangle with thickness 2.

```python
for (x,y,w,h) in total_faces:
     cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

Finally We have to display the image in which faces are bounded by a rectangle.

```python
plt.imshow(image)
```

## Saving the image

```python
cv2.imwrite('image1.jpg',image)
```

Now Let us create a generalised function for the entire face detection process.

```python
def Face_Detection(image):
    image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    total_faces = face_haar_cascade.detectMultiScale(gray_image, scaleFactor = 1.2, minNeighbors = 5)
    for (x,y,w,h) in total_faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    return image
```

## Let us check for multiple faces Image

```python
image_2=cv2.imread('actors.jpeg')
```

```python
plt.imshow(Face_Detection(image_2))
```
